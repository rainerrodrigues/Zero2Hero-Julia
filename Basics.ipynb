# Babylonian

```julia
function babylonian(y,ϵ,x0=1)
    steps = 0
    x = x0
    
    while abs(x^2 - y) > ϵ
        x = 0.5*(x + y/x)
        steps += 1
    end
    steps
end
```

```julia
babylonian(1000,0.25)
```

```julia
function nucleotideCount(strand)
    counter = Dict(base => 0 for base in "ACGT")
    for base in strand
        if (haskey(counter,base))
            counter[base] += 1
        else
            throw(Exception("Invalid strand"))
        end
    end
    counter
end
    
```

```julia
s = "AGAGAGATCCCTTA"
nucleotideCount(s)
```

```julia
function fiboSeries(n)
    if n <= 1
        return n
    else
        return fiboSeries(n-1) + fiboSeries(n-2)
    end
end

num = collect(1:8)
map(fiboSeries,num)
```

```julia
function HammingDist(strand1,strand2)
    return count(x -> x[1]!= x[2],zip(strand1,strand2))
end

HammingDist("ATC","CAT")
```

```julia
using CairoMakie

# Create a figure
fig = Figure()

# Define the coordinates for the boxes
boxes = [
    (1, 1:2, "blue"),
    (2:3, 1, "purple"),
    (1:2, 3, "brown"),
    (2, 2, "green"),
    (3, 2:3, "coral")
]

# Create boxes
for (rows, cols, color) in boxes
    Box(fig[rows, cols], color=color, strokewidth=0)
end

# Define parameters for the spiral
n = 4
t = 0:0.01:2n*π
x = (2n*π .- t) .* cos.(t)
y = (2n*π .- t) .* sin.(t)

# Create an axis
a1 = Axis(fig[:, :])

# Plot the spiral
lines!(a1, x, y, color="black")

# Hide decorations
hidedecorations!(a1, grid=false, label=false)
hidespines!(a1)

# Show the plot
# show(fig)
fig
```

```julia
#Collatz Conjecture program

function collatz_counter(n)
    if n <= 0
        throw(ArgumentError("Input must be a positive integer"))
    end
    
    steps = 0
    while n != 1
        if n % 2 == 0
            n = div(n, 2)
        else
            n = 3n + 1
        end
        steps += 1
    end
    return steps
end
```

```julia
collatz_counter(5)
```

```julia
collatz_counter(100)
```

```julia
function henon_map(u0, N)
    x0, y0 = u0
    result = [(x0, y0) for _ in 1:N]  # Create a list to store the results
    
    for i in 2:N
        x, y = result[i-1]
        x1 = 1 - 1.4x^2 + y
        y1 = 0.3x
        result[i] = (x1, y1)
    end
    
    return result
end
```

```julia
u0 = (0.0,0.0)
henon_map(u0,4)
```
